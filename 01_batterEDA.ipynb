```python
import pandas as pd
```

```python
df = pd.read_csv('./batter-2022.csv')
df.head()
```

## Column 설명
| 컬럼명 | 의미 |
| :--: | -- |
|`last_name` |  선수의 성씨 (Last Name) |
|`first_name` | 선수의 이름 (First Name) |
|`player_id` | 선수를 고유하게 식별하는 Player ID|
|`attempts` | 타석 수 또는 타구 시도 횟수 (Number of Plate Appearances or At-bat Attempts)|
|`avg_hit_angle` | 평균 타구 각도 (Average Hit Angle) - 타구가 어떤 각도로 나가는지 평균값|
|`anglesweetspotpercent` | 스윗스팟에서의 타구 비율 (Angle Sweet Spot Percent) - 스윗스팟에서의 타구 비율(%)|
|`max_hit_speed` | 최대 타구 속도 (Maximum Hit Speed) - 타구 중 가장 빠른 속도|
|`avg_hit_speed` | 평균 타구 속도 (Average Hit Speed) - 모든 타구의 평균 속도|
|`fbld` | 플라이 볼 + 직구 땅볼 타구 속도 (Fly Ball + Line Drive Ball Exit Speed) - 플라이 볼과 직구 땅볼의 타구 속도|
|`gb` | 땅볼 타구 속도 (Ground Ball Exit Speed) - 땅볼의 타구 속도|
|`max_distance` | 최대 타구 비거리 (Maximum Batted Ball Distance) - 타구 중 가장 멀리 나간 거리|
|`avg_distance` | 평균 타구 비거리 (Average Batted Ball Distance) - 모든 타구의 비거리 평균|
|`avg_hr_distance`: | 평균 홈런 비거리 (Average Home Run Distance) - 모든 홈런 타구의 비거리 평균|
|`ev95plus` | 95mph 이상의 타구 속도 비율 (Exit Velocity 95mph and Above) - 95mph 이상의 타구 비율(%)|
|`ev95percent` | 타구 속도 95mph 이상 비율 (Exit Velocity 95mph and Above Percent) - 타구 중 95mph 이상의 속도를 가진 타구 비율(%)|
|`barrels` | Barrel 타구 수 (Barrels) - Barrel로 분류되는 타구의 수|
|`brl_percent` | Barrel 타구 비율 (Barrel Percent) - Barrel 타구 비율(%)|
|`brl_pa` | Plate Appearance 당 Barrel 타구 비율 (Barrels per Plate Appearance) - 타석 당 Barrel 타구 비율|
